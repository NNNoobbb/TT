```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your Iris CSV
iris = pd.read_csv(r"C:\ThirdYear_Notes\Practicals\DSBDA\IRIS.csv")

# 1. Feature types
print("Feature types:\n", iris.dtypes, "\n")

# 2. Histograms for each numeric feature
iris.hist(figsize=(10, 8))
plt.suptitle('Histograms of Iris Features')
plt.show()

# 3. Box plots for each numeric feature
plt.figure(figsize=(10, 8))
plt.suptitle('Box plots for each numeric feature')
for i, col in enumerate(iris.columns[:4], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=iris[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```
