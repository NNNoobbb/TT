```python
# 1. Import and load
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

emp1 = pd.read_csv(r"C:\ThirdYear_Notes\Practicals\DSBDA\3_employees.csv")

# 2. Missing values & stats
print(emp1.isnull().sum())
print(emp1.describe(include='all'))

# 3. Variable summary & dimensions
summary1 = pd.DataFrame({
    'dtype': emp1.dtypes,
    'unique': emp1.nunique(),
    'non-null': emp1.count()
})
print(summary1)
print("Shape:", emp1.shape)

# 4. Type conversions example
if 'EmployeeID' in emp1.columns:
    emp1['EmployeeID'] = emp1['EmployeeID'].astype(str)
# parse dates if any
for col in emp1.select_dtypes('object'):
    try:
        emp1[col] = pd.to_datetime(emp1[col])
    except:
        pass

# 5. Z-score normalization + boxplot on Salary
if 'Salary' in emp1.columns:
    emp1['Salary_z'] = zscore(emp1['Salary'].dropna())
    sns.boxplot(x=emp1['Salary'])
    plt.title("Salary Boxplot")
    plt.show()

# 6. Encode categoricals
emp1_enc = pd.get_dummies(emp1, drop_first=True)
emp1_enc.head()
```
