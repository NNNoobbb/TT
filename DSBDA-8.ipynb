```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Analytics I: Predict Salary from Years of Experience
from sklearn.linear_model import LinearRegression

# 1. Load the salary data
salary_df = pd.read_csv(r"C:\ThirdYear_Notes\Practicals\DSBDA\Salary_Data.csv")
#    Expected columns: 'Experience', 'Salary'
print(salary_df)

# 2. Prepare X and y
X = salary_df[['YearsExperience']]  
y = salary_df['Salary']      

# 3. Fit Linear Regression model
model = LinearRegression()
model.fit(X, y)

# 4. Display coefficients
print(f"Coefficient (slope): {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")

# 5. Predict salary for 15 years of experience
years = [[15]]
predicted_salary = model.predict(years)[0]
print(f"Predicted salary for 15 years' experience: {predicted_salary:.2f}")

# 6. Plot the data and fitted line
plt.scatter(X, y, label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Fitted Line')
plt.scatter([15], [predicted_salary], color='green', marker='X', s=100,
            label="Prediction (15 yrs)")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Salary vs. Experience')
plt.legend()
plt.show()
```
